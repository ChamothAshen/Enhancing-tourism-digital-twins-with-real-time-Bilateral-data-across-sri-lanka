```python
from ultralytics import YOLO
import cv2

# Load YOLOv8n
model = YOLO('yolov8n.pt')

# Read frame
frame = cv2.imread('people.jpg')

# Inference
results = model(frame, classes=[0])  # class 0 = person

# Draw and count
count = 0
for r in results:
    boxes = r.boxes
    count += len(boxes)

print("People:", count)
```

```python
from ultralytics import YOLO
import cv2

# Load YOLOv8 model
model = YOLO("best.pt")

# Read image
image_path = "seq_000003.jpg"
frame = cv2.imread(image_path)

# Run inference (class 0 = person)
results = model(frame, conf=0.4, classes=[0])

count = 0

# Iterate over detections
for r in results:
    boxes = r.boxes
    for box in boxes:
        count += 1

        # Get box coordinates
        x1, y1, x2, y2 = map(int, box.xyxy[0])

        # Confidence
        conf = float(box.conf[0])

        # Draw rectangle
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Label
        label = f"Person {conf:.2f}"
        cv2.putText(
            frame,
            label,
            (x1, y1 - 8),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.5,
            (0, 255, 0),
            2
        )

# Put total count on image
cv2.putText(
    frame,
    f"Total People: {count}",
    (20, 40),
    cv2.FONT_HERSHEY_SIMPLEX,
    1,
    (0, 0, 255),
    3
)

print("People:", count)

# Show image
cv2.imshow("Crowd Detection", frame)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Save output
cv2.imwrite("crowd_output.jpg", frame)
```
