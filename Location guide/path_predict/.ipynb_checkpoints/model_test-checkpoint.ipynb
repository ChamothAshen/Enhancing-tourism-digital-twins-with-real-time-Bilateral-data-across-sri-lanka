```python
import os
import math
import json
import random
import joblib
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier

RNG = np.random.default_rng(42)
random.seed(42)

# -----------------------------
# 1) POI definitions (Sigiriya area)
# -----------------------------
def make_sigiriya_pois():
    # You can change these freely; keep names stable (used as classes)
    pois = [
        {"poi": "Entrance", "lat": 7.9569, "lon": 80.7600, "mean_dwell": 10},
        {"poi": "Ticket_Counter", "lat": 7.9572, "lon": 80.7604, "mean_dwell": 12},
        {"poi": "Water_Gardens", "lat": 7.9582, "lon": 80.7605, "mean_dwell": 25},
        {"poi": "Boulder_Gardens", "lat": 7.9590, "lon": 80.7610, "mean_dwell": 20},
        {"poi": "Mirror_Wall", "lat": 7.9601, "lon": 80.7613, "mean_dwell": 15},
        {"poi": "Frescoes", "lat": 7.9610, "lon": 80.7612, "mean_dwell": 18},
        {"poi": "Lion_Paws", "lat": 7.9620, "lon": 80.7610, "mean_dwell": 10},
        {"poi": "Summit", "lat": 7.9630, "lon": 80.7607, "mean_dwell": 35},
        {"poi": "Museum", "lat": 7.9564, "lon": 80.7596, "mean_dwell": 20},
        {"poi": "Exit", "lat": 7.9560, "lon": 80.7592, "mean_dwell": 5},
    ]
    return pd.DataFrame(pois)

def haversine_m(lat1, lon1, lat2, lon2):
    # meters
    R = 6371000.0
    phi1, phi2 = math.radians(lat1), math.radians(lat2)
    dphi = math.radians(lat2 - lat1)
    dl = math.radians(lon2 - lon1)
    a = math.sin(dphi/2)**2 + math.cos(phi1)*math.cos(phi2)*math.sin(dl/2)**2
    return 2*R*math.asin(math.sqrt(a))

# -----------------------------
# 2) Build a transition model for synthetic users
# -----------------------------
def build_base_transition(pois):
    """
    Create a plausible transition probability matrix (Markov-like).
    We'll make it mostly "forward path" Entrance -> ... -> Summit -> Exit,
    plus some detours to Museum and occasional skips.
    """
    poi_names = pois["poi"].tolist()
    idx = {p:i for i,p in enumerate(poi_names)}
    n = len(poi_names)
    T = np.zeros((n, n), dtype=float)

    def add(a, b, w):
        T[idx[a], idx[b]] += w

    # Main forward route
    route = ["Entrance","Ticket_Counter","Water_Gardens","Boulder_Gardens","Mirror_Wall",
             "Frescoes","Lion_Paws","Summit","Exit"]
    for i in range(len(route)-1):
        add(route[i], route[i+1], 0.70)
        # sometimes skip one stop
        if i+2 < len(route):
            add(route[i], route[i+2], 0.15)
        # sometimes detour to museum early
        if route[i] in ["Entrance","Ticket_Counter","Water_Gardens"]:
            add(route[i], "Museum", 0.10)
        # small chance to go backward / wander
        if i-1 >= 0:
            add(route[i], route[i-1], 0.05)

    # Museum usually goes back to route or exit
    add("Museum", "Ticket_Counter", 0.35)
    add("Museum", "Water_Gardens", 0.35)
    add("Museum", "Exit", 0.30)

    # Summit mostly goes to Exit
    add("Summit", "Exit", 0.90)
    add("Summit", "Lion_Paws", 0.10)

    # Exit is terminal (self-loop)
    add("Exit", "Exit", 1.0)

    # Normalize rows
    for i in range(n):
        s = T[i].sum()
        if s == 0:
            T[i] = 1.0 / n
        else:
            T[i] /= s
    return T, poi_names

def sample_next_poi(T, poi_names, current_poi, crowd, remaining_minutes):
    """
    Modify probabilities based on crowd + remaining time:
    - if crowd high, more likely to skip forward / exit
    - if remaining time low, more likely to exit
    """
    idx = {p:i for i,p in enumerate(poi_names)}
    cur_i = idx[current_poi]
    probs = T[cur_i].copy()

    # Push toward "Exit" if time is low
    exit_i = idx["Exit"]
    if remaining_minutes < 45:
        probs[exit_i] += 0.25
    if remaining_minutes < 25:
        probs[exit_i] += 0.50

    # High crowd -> skip ahead + exit
    if crowd > 0.75:
        probs[exit_i] += 0.20

    # Normalize
    probs = np.clip(probs, 0, None)
    probs = probs / probs.sum()
    return RNG.choice(poi_names, p=probs)

def crowd_level(hour):
    """
    Simple crowd curve: busier mid-morning to afternoon.
    """
    base = 0.2
    if 9 <= hour <= 15:
        base = 0.6
    if 10 <= hour <= 13:
        base = 0.8
    noise = RNG.normal(0, 0.08)
    return float(np.clip(base + noise, 0, 1))

def simulate_session(pois_df, T, poi_names, user_id, session_id):
    """
    Create one user session with time budget 120..240 minutes.
    """
    poi_meta = {r["poi"]: r for _, r in pois_df.iterrows()}
    total_minutes = int(RNG.integers(120, 241))   # user input later: 2h/3h/4h
    start_hour = int(RNG.integers(7, 16))         # visiting hours
    day_of_week = int(RNG.integers(0, 7))

    current_poi = "Entrance"
    remaining = total_minutes
    step = 0
    rows = []

    while step < 25 and remaining > 0:
        hour_now = int(np.clip(start_hour + (total_minutes - remaining)//60, 0, 23))
        crowd = crowd_level(hour_now)

        # dwell at current poi (affected by crowd: crowd high -> dwell shorter)
        mean_dwell = poi_meta[current_poi]["mean_dwell"]
        dwell = max(2, int(RNG.normal(mean_dwell * (1.05 - 0.5*crowd), 4)))

        # If can't afford dwell, end session
        if remaining - dwell <= 0:
            break    

        # choose next
        next_poi = sample_next_poi(T, poi_names, current_poi, crowd, remaining)

        # walking minutes based on distance
        cur = poi_meta[current_poi]
        nxt = poi_meta[next_poi]
        dist_m = haversine_m(cur["lat"], cur["lon"], nxt["lat"], nxt["lon"])
        # walking speed ~ 70 m/min (slow + crowds)
        walk = max(1, int(dist_m / 70.0 + RNG.normal(0, 1.2)))
        walk = int(np.clip(walk, 1, 30))

        # If next is same (Exit self-loop), keep minimal walk
        if current_poi == "Exit":
            next_poi = "Exit"
            walk = 1

        # Update remaining
        remaining_after = remaining - (dwell + walk)
        if remaining_after < 0:
            break

        rows.append({
            "user_id": int(user_id),
            "session_id": int(session_id),
            "step_idx": int(step),
            "hour": int(hour_now),
            "day_of_week": int(day_of_week),
            "current_poi": current_poi,
            "current_lat": float(cur["lat"]),
            "current_lon": float(cur["lon"]),
            "current_crowd": float(crowd),
            "remaining_minutes": int(remaining),
            "next_poi": next_poi,
            "dwell_minutes": int(dwell),
            "walk_minutes": int(walk),
        })

        remaining = remaining_after
        current_poi = next_poi
        step += 1

        if current_poi == "Exit":
            break

    return rows

def generate_synthetic_movement_dataset(n_users=400, sessions_per_user=2):
    pois = make_sigiriya_pois()
    T, poi_names = build_base_transition(pois)

    all_rows = []
    session_id = 0
    for u in range(n_users):
        for _ in range(sessions_per_user):
            all_rows.extend(simulate_session(pois, T, poi_names, user_id=u, session_id=session_id))
            session_id += 1

    df = pd.DataFrame(all_rows)
    return df, pois

# -----------------------------
# 3) Train 3 models to predict next_poi
# -----------------------------
def top_k_accuracy(y_true, proba, classes, k=3):
    """
    proba: (n_samples, n_classes)
    """
    true_idx = np.array([np.where(classes == y)[0][0] for y in y_true])
    topk = np.argsort(proba, axis=1)[:, -k:]
    hits = [true_idx[i] in topk[i] for i in range(len(true_idx))]
    return float(np.mean(hits))

def train_and_compare(df, save_dir="saved_models"):
    os.makedirs(save_dir, exist_ok=True)

    feature_cols = [
        "hour", "day_of_week", "current_poi", "current_lat", "current_lon",
        "current_crowd", "remaining_minutes"
    ]
    target_col = "next_poi"

    X = df[feature_cols].copy()
    y = df[target_col].copy()

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42, stratify=y
    )

    cat_cols = ["current_poi"]
    num_cols = ["hour", "day_of_week", "current_lat", "current_lon", "current_crowd", "remaining_minutes"]

    pre = ColumnTransformer(
        transformers=[
            ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
            ("num", StandardScaler(), num_cols),
        ]
    )

    models = {
        "logreg_multinomial": LogisticRegression(
            max_iter=2000, n_jobs=None
        ),
        "random_forest": RandomForestClassifier(
            n_estimators=350, max_depth=None, min_samples_leaf=2, random_state=42, n_jobs=-1
        ),
        "hist_gb": HistGradientBoostingClassifier(
            max_depth=6, learning_rate=0.08, random_state=42
        ),
    }

    results = []
    fitted = {}

    for name, model in models.items():
        pipe = Pipeline([("pre", pre), ("model", model)])
        pipe.fit(X_train, y_train)

        y_pred = pipe.predict(X_test)
        acc = accuracy_score(y_test, y_pred)

        # Top-3 needs predict_proba (HistGB has it; RF has it; LogReg has it)
        proba = pipe.predict_proba(X_test)
        classes = pipe.named_steps["model"].classes_
        top3 = top_k_accuracy(y_test.to_numpy(), proba, classes, k=3)

        joblib.dump(pipe, os.path.join(save_dir, f"{name}.joblib"))

        results.append({"model": name, "top1_acc": acc, "top3_acc": top3})
        fitted[name] = pipe

        print(f"[{name}] top-1 acc={acc:.3f} | top-3 acc={top3:.3f}")

    results_df = pd.DataFrame(results).sort_values("top1_acc", ascending=False)
    results_df.to_csv(os.path.join(save_dir, "results.csv"), index=False)

    best_name = results_df.iloc[0]["model"]
    print("\nBest model:", best_name)

    meta = {
        "feature_cols": feature_cols,
        "target_col": target_col,
        "best_model": best_name
    }
    with open(os.path.join(save_dir, "meta.json"), "w") as f:
        json.dump(meta, f, indent=2)

    return results_df, (X_test, y_test)

# -----------------------------
# 4) Load saved models + test function
# -----------------------------
def load_models_and_test(save_dir, X_test, y_test):
    """
    Loads all .joblib models in save_dir and prints top-1/top-3 again.
    """
    files = [f for f in os.listdir(save_dir) if f.endswith(".joblib")]
    if not files:
        raise FileNotFoundError("No saved models found in " + save_dir)

    for f in sorted(files):
        name = f.replace(".joblib", "")
        pipe = joblib.load(os.path.join(save_dir, f))

        y_pred = pipe.predict(X_test)
        acc = accuracy_score(y_test, y_pred)

        proba = pipe.predict_proba(X_test)
        classes = pipe.named_steps["model"].classes_
        top3 = top_k_accuracy(y_test.to_numpy(), proba, classes, k=3)

        print(f"[LOAD TEST {name}] top-1 acc={acc:.3f} | top-3 acc={top3:.3f}")

# -----------------------------
# 5) Recommender: itinerary builder (time-budget aware)
# -----------------------------
def recommend_itinerary(current_poi, start_hour, hours_available, model, pois_df, day_of_week=2, max_stops=6):
    """
    Creates a recommended list of POIs starting from current_poi within a time budget.
    """
    poi_meta = {r["poi"]: r for _, r in pois_df.iterrows()}
    remaining = int(hours_available * 60)

    itinerary = []
    step = 0
    cur = current_poi

    while step < max_stops and remaining > 10 and cur != "Exit":
        hour_now = int(np.clip(start_hour + (hours_available*60 - remaining)//60, 0, 23))
        crowd = crowd_level(hour_now)

        cur_meta = poi_meta[cur]
        X_one = pd.DataFrame([{
            "hour": hour_now,
            "day_of_week": int(day_of_week),
            "current_poi": cur,
            "current_lat": float(cur_meta["lat"]),
            "current_lon": float(cur_meta["lon"]),
            "current_crowd": float(crowd),
            "remaining_minutes": int(remaining),
        }])
      
        proba = model.predict_proba(X_one)[0]
        classes = model.named_steps["model"].classes_
        # pick best next (avoid staying same unless Exit)
        ranked = classes[np.argsort(proba)[::-1]].tolist()

        next_poi = None
        for cand in ranked:
            if cand != cur:
                next_poi = cand
                break
        if next_poi is None:
            next_poi = "Exit"

        # estimate time cost
        dwell = max(2, int(RNG.normal(poi_meta[cur]["mean_dwell"], 4)))
        dist_m = haversine_m(cur_meta["lat"], cur_meta["lon"], poi_meta[next_poi]["lat"], poi_meta[next_poi]["lon"])
        walk = max(1, int(dist_m / 70.0))

        cost = dwell + walk
        if remaining - cost < 0:
            break

        itinerary.append({
            "from": cur,
            "to": next_poi,
            "hour": hour_now,
            "est_dwell_min": dwell,
            "est_walk_min": walk,
            "remaining_before": remaining
        })

        remaining -= cost
        cur = next_poi
        step += 1

        if cur == "Exit":
            break

    return itinerary
```

```python
df, pois = generate_synthetic_movement_dataset(n_users=600, sessions_per_user=2)
df.to_csv("synthetic_movement.csv", index=False)
print("Saved dataset: synthetic_movement.csv")
print("Rows:", len(df), " Unique next_poi:", df["next_poi"].nunique())

results_df, (X_test, y_test) = train_and_compare(df, save_dir="saved_models")
print("\nModel comparison:\n", results_df)

print("\nReload + test saved models:")
load_models_and_test("saved_models", X_test, y_test)

# Example: build itinerary using the best model
meta = json.load(open("saved_models/meta.json"))
best_model = joblib.load(os.path.join("saved_models", f"{meta['best_model']}.joblib"))

example = recommend_itinerary(
    current_poi="Water_Gardens",
    start_hour=10,
    hours_available=2.5,
    model=best_model,
    pois_df=pois,
    day_of_week=5,
    max_stops=6
)
print("\nExample itinerary (2.5 hours):")
for item in example:
    print(item)
```
